## Text Preprocessing

### Subtask:
Perform text cleaning, including steps like lowercasing, removing punctuation, special characters, and numbers. Then, tokenize the words, remove stopwords, apply lemmatization, and finally rejoin the processed tokens into clean text.


**Reasoning**:
First, I will download the necessary NLTK data (stopwords and wordnet) required for text preprocessing tasks like stopword removal and lemmatization. Then, I will import NLTK and other required libraries.



```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

print("NLTK downloads and imports complete.")
```

**Reasoning**:
Now that the necessary NLTK packages are downloaded and imported, I will proceed with the text cleaning steps as outlined in the instructions: lowercasing, removing punctuation, special characters, and numbers, tokenization, stopword removal, lemmatization, and finally rejoining the words into a clean text column.



```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # 3. Remove special characters and numbers (keeping only alphabetic characters and spaces)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 4. Tokenize words
    tokens = nltk.word_tokenize(text)

    # 5. Remove stopwords and 6. Apply lemmatization
    processed_tokens = []
    for word in tokens:
        if word not in stop_words:
            processed_tokens.append(lemmatizer.lemmatize(word))

    # 7. Rejoin words
    return ' '.join(processed_tokens)

df['processed_text'] = df['News'].apply(preprocess_text)
display(df.head())
```

## Create Bag-of-Words Model

### Subtask:
Generate a Bag-of-Words (BoW) representation of the preprocessed text. Then, convert the BoW model into a pandas DataFrame for further analysis. This step can also be modified later to use TF-IDF if desired.


**Reasoning**:
To create a Bag-of-Words representation, I will import `CountVectorizer`, initialize it, fit it to the preprocessed text, transform the text into a document-term matrix, and then convert this matrix into a pandas DataFrame for inspection.



```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer
# You can adjust min_df and max_df to filter words, e.g., min_df=2 to ignore words that appear in less than 2 documents
vectorizer = CountVectorizer()

# Fit and transform the 'processed_text' column to create the Document-Term Matrix (DTM)
dtm = vectorizer.fit_transform(df['processed_text'])

# Create a pandas DataFrame from the DTM
df_dtm = pd.DataFrame(dtm.toarray(), columns=vectorizer.get_feature_names_out())

# Display the first few rows and the shape of the DTM DataFrame
print("Bag-of-Words DataFrame Head:")
display(df_dtm.head())
print("\nShape of the Bag-of-Words DataFrame:", df_dtm.shape)
```

## Apply NMF for Topic Modeling

### Subtask:
Apply Non-negative Matrix Factorization (NMF) to the Bag-of-Words DataFrame to perform topic modeling and identify underlying themes in the text data.


**Reasoning**:
To apply Non-negative Matrix Factorization (NMF) for topic modeling, I need to import the NMF class, instantiate it with a specified number of topics, and then fit the model to the Bag-of-Words DataFrame to generate the document-topic and topic-word matrices.



```python
from sklearn.decomposition import NMF

# Instantiate NMF model with 2 components (topics)
n_components = 2
nmf_model = NMF(n_components=n_components, random_state=42, init='nndsvda')

# Fit the NMF model to the Bag-of-Words DataFrame
# W (document-topic) matrix
W = nmf_model.fit_transform(df_dtm)
# H (topic-word) matrix
H = nmf_model.components_

print(f"NMF model fitted with {n_components} components.")
print(f"Shape of W (Document-Topic Matrix): {W.shape}")
print(f"Shape of H (Topic-Word Matrix): {H.shape}")
```

**Reasoning**:
Now that the NMF model has been fitted and the topic-word matrix (H) is available, I will extract and display the top keywords for each identified topic. This will provide insight into the themes represented by each topic, which is a key part of the overall task.



```python
feature_names = vectorizer.get_feature_names_out()
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx+1}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
    print("\nTop keywords for each topic displayed.")

no_top_words = 3 # Number of top keywords to display for each topic
display_topics(nmf_model, feature_names, no_top_words)
```
